```python
import numpy as np
from PIL import Image
import glob
images = []
resized = []
resized_array = []
for pictures in glob.glob(r'food/*.jpg'):
    images.append(Image.open(pictures))
for image in images:
    resized.append(image.resize([200,200]))
for items in resized:
    numpy = np.array(items)
    resized_array.append(numpy)
print(resized_array,'\n',np.shape(resized_array))
```
